# 03 - Exploratory Data Analysis (EDA)
This notebook performs comprehensive EDA and saves publication-quality visualization charts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import os

# Setup plotting style
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

conn = sqlite3.connect('../data/db/bluestock_mf.db')
df_nav = pd.read_sql('SELECT * FROM fact_nav', conn)
df_funds = pd.read_sql('SELECT * FROM dim_fund', conn)
df_tx = pd.read_sql('SELECT * FROM fact_transactions', conn)
df_sip = pd.read_sql('SELECT * FROM fact_sip_industry', conn)

os.makedirs('../reports/charts', exist_ok=True)
print('Data Loaded successfully for EDA.')
```

### Chart 1: NAV Trends of Top Schemes

```python
for code, group in df_nav.groupby('amfi_code'):
    name = df_funds[df_funds['amfi_code'] == code]['scheme_name'].values[0]
    plt.plot(pd.to_datetime(group['date']), group['nav'], label=name[:20])
plt.title('NAV Trend Analysis (2022-2026)')
plt.xlabel('Date')
plt.ylabel('NAV')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('../reports/charts/nav_trends.png')
plt.show()
```

### Chart 2: Transaction Type distribution

```python
df_tx['transaction_type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#4F46E5', '#10B981', '#EF4444'])
plt.title('Investor Transaction Type Split')
plt.ylabel('')
plt.savefig('../reports/charts/transaction_distribution.png')
plt.show()
```
